```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt

x1=np.linspace(-2,2,100)
y1=x1
x2,y2=np.meshgrid(x1,y1)
z=x2**2+x2*y2+y2**3 -2
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(x2,y2,z)

z2=3*x2+4*y2-6
ax.plot_surface(x2,y2,z2)

t=x1
x3=1-3*t
y3=1-4*t
z3=1+t
ax.plot3D(x3,y3,z3)
plt.show()
```
