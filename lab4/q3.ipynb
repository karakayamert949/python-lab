```python
import numpy as np
import matplotlib.pyplot as plt

ph=plt.imread("photo.png")
ph.astype('uint8')
p3=ph.copy()
ph*=255
p2=ph.copy()

rph=ph[:,:,0]
gph=ph[:,:,1]
bph=ph[:,:,2]

rph[np.where(rph<127)]=0
rph[np.where(rph>=127)]=255

gph[gph<127]=0
gph[gph>=127]=255

bph[bph<127]=0
bph[bph>=127]=255

p2[:,:,0]=rph
p2[:,:,1]=gph
p2[:,:,2]=bph

fig=plt.figure(figsize=(8,6))
fig.add_subplot(1,2,1)
plt.imshow(p3)
fig.add_subplot(1,2,2)
plt.imshow(p2)
```
