```python
import numpy as np
import skimage.io as io
import matplotlib.pyplot as plt

photo=io.imread("photo.png")
x,y,z=photo.shape
topl=photo[0:x//2,0:y//2,:]
topr=photo[0:x//2,y//2:,:]
bot=photo[x//2:,0:,:]

ml=[]
ml.append(np.mean(topl[:,:,0]))
ml.append(np.mean(topl[:,:,1]))
ml.append(np.mean(topl[:,:,2]))

mr=[]
mr.append(np.mean(topr[:,:,0]))
mr.append(np.mean(topr[:,:,1]))
mr.append(np.mean(topr[:,:,2]))

mb=[]
mb.append(np.mean(bot[:,:,0]))
mb.append(np.mean(bot[:,:,1]))
mb.append(np.mean(bot[:,:,2]))

nph=photo.copy()
nph[0:x//2,0:y//2,0]=ml[0]
nph[0:x//2,0:y//2,1]=ml[1]
nph[0:x//2,0:y//2,2]=ml[2]

nph[0:x//2,y//2:,:]=mr
nph[x//2:,0:,:]=mb

fig=plt.figure(figsize=(10,6))
fig.add_subplot(1,2,1)
plt.imshow(photo)
plt.axis('off')
plt.title("First")

fig.add_subplot(1,2,2)
plt.imshow(nph)
plt.axis('off')
plt.title("second")
```
