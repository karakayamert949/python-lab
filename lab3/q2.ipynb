```python
import numpy as np
import matplotlib.pyplot as plt

t=np.linspace(-2*np.pi,2*np.pi,3000)
alf=3
x=4*alf*((1-3*t**2)/(1+t**2)**3)
y=4*alf*((t*(3-t**2))/(1+t**2)**3)
plt.plot(x,y)
```

```python
b=2.2
k=0.5
x1=b*(t*np.cos(t)+k*t)
y1=b*t*np.sin(t)
plt.plot(x,y)
plt.plot(x1,y1)
plt.xlabel("x axis")
plt.legend(["cayley","doppler"])
```
