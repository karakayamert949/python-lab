```python
import numpy as np
import matplotlib.pyplot as plt

th=np.pi*30/180
v0=40
g=9.8
thit=np.sin(th)*2*v0/g
t=np.linspace(0,thit,100)

h=v0*t*np.sin(th)-(1/2)*g*t**2
v=np.sqrt(v0**2-2*v0*g*t*np.sin(th)+g**2 *t**2)


plt.plot(t,h)
plt.plot(t,v)
plt.legend(["height","velocity"])
t[h==(max(h))]
index=np.where((h<5) | (v>35))
print(f"times when height is less than 5 or height is greater than 35: {t[index]}")
```
