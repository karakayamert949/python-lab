```python
import numpy as np
import matplotlib.pyplot as plt

t=[]
t.append(0)
d=[]
d.append(16)
i=0
x=[]
y=[]
while d[int(t[i]*100)] > 15:
    x.append(5*t[i]-10)
    y.append(25*t[i]**2 - 120*t[i] + 144)
    d.append(np.sqrt(x[int(t[i]*100) -1]**2 +y[int(t[i]*100)-1]**2))
    t.append(t[i]+0.01)
    i+=1
d.pop(0)
t.pop(-1)
plt.plot(t,d)
plt.title("başlık")
plt.xlabel("time")
plt.ylabel("distance")

```
