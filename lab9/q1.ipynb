```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,11,100)
h=np.array([1, 0])
w=np.array([-2,22])
l=np.array([-2,40])
temp=np.polymul(h,w)
p=np.polymul(temp,l)
V=[np.polyval(p,i) for i in x]
# V=(40-2*x)*(22-2*x)*x
plt.plot(x,V)

p1000=p
p1000[-1]-=1000
a=np.roots(p1000)
a=a[((11>a) & (a>0))]
print(a)
pD=np.polyder(p)
r=np.roots(pD)
r=r[(r<11) & (r>0)]
print(r)
```
