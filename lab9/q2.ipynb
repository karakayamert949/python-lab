```python
from msilib import MSIMODIFY_INSERT_TEMPORARY
import numpy as np
import matplotlib.pyplot as plt


x=np.arange(260,310,1)
t=np.arange(273,299,5)
p=np.array([4.579,6.543,9.209,12.788,17.535,23.756])
p2=np.polyfit(t,p,2)
polyVal=[np.polyval(p2,i) for i in x]
polyVals=[np.polyval(p2,i) for i in t]
plt.plot(x,polyVal)


pE=np.polyfit(t,np.log(p),1)
expoVal=[(np.exp(pE[1])*np.exp(pE[0]*i)) for i in x]
expoVals=[(np.exp(pE[1])*np.exp(pE[0]*i)) for i in t]
print(f"{p2}\n{pE}")
plt.plot(x,expoVal)
plt.plot(t,p,'o')

MAPEP=100/6 * sum(abs(1-polyVals/p))
MAPEE=100/6 * sum(abs(1-expoVals/p))

if MAPEE<MAPEP:
    print(f"the value of P is {np.polyval(p2,300)}, for T=300")
else:
    print(f"the value of P is {np.exp(pE[1])*np.exp(pE[0]*300)}, for T=300")
```
