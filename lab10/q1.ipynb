```python
import sympy as sym
import numpy as np
from sympy.solvers import solve
import matplotlib.pyplot as plt

x=sym.Symbol("x")
y=sym.Symbol("y")
A=sym.Matrix([[0,1,3],[2,x,4],[y,1,7]])
print(A)

x1=np.arange(-5,1,0.1)
s=sym.det(A)
y1=[]
s2=s.subs(x,x1[2])
a=solve(s2,y)
print(a)
for i in x1:
    s2=s.subs(x,i)
    y1.append(solve(s2,y))

plt.plot(x1,y1,"o")
```

```python
y=sym.Symbol("y")
x=sym.Symbol("x")

y=(3/2)*(x-3)**2 +1
sym.plot(y,(x,0,6),(y,0,7))
inters=sym.solve(y-4,x)
x1=min(inters)
x2=max(inters)
t=sym.integrate(y,(x,x1,x2))
area=4*(x2-x1)-t
print(area)
```
