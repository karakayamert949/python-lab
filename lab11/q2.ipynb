```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate

x=np.arange(0,6)
y=np.array([1,-0.6242,-1.4707,3.2406,-0.7366,-6.3717])

xd=np.arange(0,5.1,0.1)
f=(1.5**xd)*np.cos(2*xd)
f2=interpolate.interp1d(x,y)
tck=interpolate.splrep(x, y, s=0)
f3=interpolate.splev(xd,tck,der=0)
f4=interpolate.PchipInterpolator(x, y)

plt.plot(x,y,'o',xd,f,xd,f2(xd),xd,f3,xd,f4(xd))
plt.legend(["points","original","linear","spline","pchip"])
```
