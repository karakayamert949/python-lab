```python
import numpy as np
import matplotlib.pyplot as plt

w=np.arange(1,14,2)
w1=np.arange(1,13.1,0.1)
h=np.array([22,51,127,202,227,248,252])

# y = 254/h -1
y = 254/h -1

# ln(y)=ln(A*exp(-B*w))
# ln(A) -B*w

p=np.polyfit(w,np.log(y),1)
A=np.exp(p[1])
B=-p[0]

func=254/(1+A*np.exp(-B*w1))

plt.plot(w1,func,w,h,'o')
```
