```python
import numpy as np
import matplotlib.pyplot as plt
sum1=0
N=1
x=np.linspace(-np.pi,np.pi,100)
for i in range(N):
    temp=(2/np.pi)*(((-1)**(i+2))/(i+1))*np.sin((i+1)*x)
    sum1+=temp
```

```python
N=5
sum2=0
x=np.linspace(-np.pi,np.pi,100)
for i in range(N):
    temp=(2/np.pi)*(((-1)**(i+2))/(i+1))*np.sin((i+1)*x)
    sum2+=temp
```

```python
N=60
sum3=0
x=np.linspace(-np.pi,np.pi,100)
for i in range(N):
    temp=(2/np.pi)*(((-1)**(i+2))/(i+1))*np.sin((i+1)*x)
    sum3+=temp

sum4=x/np.pi
plt.plot(x,sum1)
plt.plot(x,sum2)
plt.plot(x,sum3)
plt.plot(x,sum4)
plt.legend(["N=1","N=5","N=60","Original"])
```
